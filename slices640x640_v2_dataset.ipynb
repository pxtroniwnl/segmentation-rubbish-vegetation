```python
from roboflow import Roboflow
rf = Roboflow(api_key=api_key)
project = rf.workspace("proyecto-prueba").project("solid-waste-detection-vnlff")
version = project.version(12)
dataset = version.download("yolov11")
```

```python
import os
from sahi.slicing import slice_image

# 1. Ruta de origen (lo que descargaste de Roboflow)
input_dir = "/content/solid-waste-detection-11/train/images"

# 2. Ruta de destino (EN TU DRIVE para que no se borre)
output_dir = "/content/drive/MyDrive/Dataset_Dron_640_Sliced"
os.makedirs(output_dir, exist_ok=True)
```

```python
output_base_path = "/content/drive/MyDrive/Dataset_Basura_640_Sliced"
os.makedirs(output_base_path, exist_ok=True)
```

```python
from sahi.slicing import slice_image

# Rutas de origen (lo que descargó Roboflow)
input_images_dir = f"{dataset.location}/train/images"
output_images_dir = os.path.join(output_base_path, "train/images")

os.makedirs(output_images_dir, exist_ok=True)

# Parámetros técnicos
SLICE_SIZE = 640
OVERLAP = 0.2  # 20% de solapamiento para no perder basura en los bordes

# Listar y procesar
images = [f for f in os.listdir(input_images_dir) if f.endswith(('.jpg', '.jpeg', '.png'))]

print(f"Iniciando el proceso para {len(images)} imágenes...")

for img_name in images:
    slice_image(
        image=os.path.join(input_images_dir, img_name),
        output_file_name=os.path.splitext(img_name)[0],
        output_dir=output_images_dir,
        slice_height=SLICE_SIZE,
        slice_width=SLICE_SIZE,
        overlap_height_ratio=OVERLAP,
        overlap_width_ratio=OVERLAP,
    )

print(f"-----Slicing completado. Los archivos están en tu Drive: {output_images_dir}-----")
```
