```python
import os
import cv2
import shutil
import numpy as np # Importante para el conteo
from tqdm import tqdm

# --- CONFIGURACIÓN DE RUTAS ---
dataset_dir = "/mnt/c/Users/Pc/Pictures/images" 
labels_dir = "/mnt/c/Users/Pc/Pictures/labels"
cleaned_dir = "/mnt/c/Users/Pc/Pictures/images-back-blackg"
cleaned_labels_dir = "/mnt/c/Users/Pc/Pictures/labels-back-blackg"

UMBRAL_NEGRO = 0.15 

os.makedirs(cleaned_dir, exist_ok=True)
os.makedirs(cleaned_labels_dir, exist_ok=True)

if not os.path.exists(dataset_dir):
    print(f"No se encuentra: {dataset_dir}")
else:
    images = [f for f in os.listdir(dataset_dir) if f.endswith(('.jpg', '.jpeg', '.png'))]
    moved_count = 0

    print(f"Analizando {len(images)} imágenes...")

    for img_name in tqdm(images):
        img_path = os.path.join(dataset_dir, img_name)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        
        if img is None:
            continue
            
        # --- CORRECCIÓN AQUÍ ---
        # Contamos cuántos píxeles son exactamente 0 (negro puro)
        black_pixels = np.sum(img == 0)
        
        total_pixels = img.shape[0] * img.shape[1]
        percentage_black = black_pixels / total_pixels
        
        if percentage_black > UMBRAL_NEGRO:
            # Mover Imagen
            shutil.move(img_path, os.path.join(cleaned_dir, img_name))
            
            # Mover Etiqueta si existe
            label_name = os.path.splitext(img_name)[0] + ".txt"
            label_src = os.path.join(labels_dir, label_name)
            if os.path.exists(label_src):
                shutil.move(label_src, os.path.join(cleaned_labels_dir, label_name))
                
            moved_count += 1

    print(f"\nProceso completado.")
    print(f"Se movieron {moved_count} imágenes y sus respectivos .txt")
```
